```python
import folium as fl
import pandas as pd
```

```python
data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
```

```python
data[['LaunchSite','Latitude','Longitude']]
```

```python
launch_map = fl.Map(location=[28.561857, -80.577366], zoom_start=12)
```

```python
# instantiate a feature group for the incidents in the dataframe
feats = fl.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(data.Latitude, data.Longitude):
    feats.add_child(
        fl.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

#
```

```python
launch_map.add_child(feats)
```
